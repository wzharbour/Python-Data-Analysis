#### 为消除数据之间量纲影响，需对数据进行规范化处理，将数据落入一个有限范围。
* 常见规范化方法将数据范围调整到[0,1] or [-1,1]。一般使用方法：
    * min-max规范化（x-min)/(max-min)
    * Z-Score规范化 (x-mean)/std
    * 小数定标规范化 x/(10^j） j表示x数据的最大绝对值的位数
    * 均值归一化 （x-mean)/(max-min)
     * replace 特殊符号的处理
     * dataframe 数据类型的转换

```python
import pandas  as pd
import numpy as np
import tushare  as ts
```

```python
#pro = ts.pro_api('eb13b3bfd2bd07fd9eb40234f19941c73f230e1e98cc212b8cd407c7')
#data = pro.daily(ts_code='000001.SZ',start_date='20180101',end_date='20201231')
data0=pd.read_excel('000027.xls')
data0['日期']=pd.to_datetime(data0['日期'])
data=pd.DataFrame(data=data0,columns=['日期', '开盘价','最高价','最低价','收盘价','涨跌额','涨跌幅(%)','成交量(手)','成交金额(万元)','振幅','换手率'])

```

```python
data1 =data.set_index('日期').sort_index()
```

```python


X = data1.iloc[:,1:]
X

```

```python
X.dtypes
```

```python
X['成交量(手)']=X['成交量(手)'].replace({',':''}, regex = True)
X['成交金额(万元)']=X['成交金额(万元)'].replace({',':''}, regex = True)
X
```

```python
X['成交量(手)']=X['成交量(手)'].astype(str).astype(float)
X['成交金额(万元)']=X['成交金额(万元)'].astype(str).astype(float)
X.dtypes
```

```python
normal_X = X.apply(lambda x : (x-x.min())/(x.max()-x.min()))
```

```python
normal_X
```

```python
normal_X_1 = X.apply(lambda x : (x-x.mean())/x.std())  # z_score
```
