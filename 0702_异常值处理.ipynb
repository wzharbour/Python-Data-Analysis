### 1. 异常值
* 经验给定某一些范围的值为异常值（如分红后的股票价格数据跳水，可以做复权处理）
* 3sigma原则确认，[均值-3sigma ，均值 + 3sigma]区间之外的值为异常值

```python
import pandas  as pd
import numpy as np
import tushare  as ts
```

```python
pro = ts.pro_api('eb13b3bfd2bd07fd9eb40234f19941c73f230e1e98cc212b8cd407c7')
#data = pro.daily(ts_code='000001.SZ',start_date='20180101',end_date='20201231')
data0=pd.read_excel('000027.xls')

data0
```

```python
data0['日期']=pd.to_datetime(data0['日期'])
data=pd.DataFrame(data=data0,columns=['日期', '开盘价','最高价','最低价','收盘价','涨跌额','涨跌幅(%)','成交量(手)','成交金额(万元)','振幅','换手率'])
#appl.set_index(keys='日期')
data = data.set_index('日期').sort_index()
data
```

##  describe()函数

* 该函数的返回值包括有均值、标准差、最大值、最小值、分位数等。
* 其中的分位数默认的分位数为[1/4、1/2、3/4]，若要重新指定，可用percentiles参数指定，如percentiles = [0.6,0.8]。

```python
data.describe()
```

```python
low = data['收盘价'].mean() - 3 * data['收盘价'].std()
high = data['收盘价'].mean() + 3 * data['收盘价'].std()
```

```python
data1 = data[~ ((data['收盘价']<low)|(data['收盘价']>high))]
```

### 2. 重复值
* 在Pandas中 duplicated()方法发现重复值（返回bool型数组,F非重复，T重复)
* drop_duplicates()方法移除重复值

```python
df = pro.query('fina_indicator', ts_code='600000.SH', 
               start_date='20170101', end_date='20201231')
df
```

```python
df[['ts_code','ann_date']].duplicated()
```

```python
df[['ts_code','ann_date']].drop_duplicates(keep='last')
```
